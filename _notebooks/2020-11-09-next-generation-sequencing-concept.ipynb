# direct sequencing(직접염기서열분석법)

## nanoball generation

![](../images/2020-11-09-ngs/Untitled23.png)

DNA Nanoball 기술은 Beijing Genomics Institute (BGI)가 가진 기술이다. 이 기술은 RCA (Rolling Circle Amplification) 방식을 통해 Library를 300-500 copies 증폭하는 기술이다. Original template만이 증폭되므로 일반적인 PCR 실험에서 발생될 수 있는 오류를 최소화하며, 오류가 한 번 발생하더라도 더 이상 축적되지 않는다. 

1. 염기서열 결정을 위한 template DNA의 양쪽 말단에 adopter를 단다. 
2. 마치 plasmid와 비슷한 모습의 circular form으로 만든다.
3. 몇 군데를 끊어 내어 끊어진 각 부위에 어댑터를 붙인다.
4. Rolling Circle Amplification(RCA)방법을 통해 PCR로 template DNA를 증폭하여 염기서열을 확보한다.

이러한 방식의 직접염기서열분석법(direct sequencing)은 분석하고자 하는 부위를 PCR로 증폭해야 하기 때문에 여러 타겟을 분석할 경우 많은 시간과 노력 및 비용이 소요되어 효율성이 낮은 문제점이 있다. 이러한 단점을 극복하고자 차세대 염기서열분석(next generation sequencing; NGS)법이 개발되었다. NGS는 DNA 가닥을 각각 하나씩 분석하는 방식으로 기존의 직접염기서열분석법에 비해 매우 빠르고 저렴하게 염기서열 분석이 가능하다는 장점을 가지고 있다. NGS는 보통 다음의 단계로 수행된다.

1. DNA를 일정한 조각(fragment)으로 분절화시킨다.
2. 장비가 인식할 수 있는 특정 염기서열을 가진 올리고뉴클레오티드(oligonucleotide)를 붙여주는 라이브러리(library)를 제작한다.
3. 각 라이브러리 DNA 가닥의 염기서열을 장비에서 읽는다.
4. 장비에서 생성된 데이터를 가공하여 알고리즘으로 분석한다.

![](../images/2020-11-09-ngs/Untitled.png)

![](../images/2020-11-09-ngs/Untitled1.png)

NGS 장비 기술의 발달로 2008년 1명의 유전체를 분석하는 데에 6개월이 걸렸다면, 2017년엔 60명의 유전체를 분석하는 데에 2일이 걸리게 되었고, 그 비용도 낮아졌다. 이는 IT의 발전과 무관하다고 볼 수 없다. 

![](../images/2020-11-09-ngs/Untitled2.png)

개인의 유전체를 분석하는 것이 왜 사업이 될까? 예를 들어 특정 화학물질이나 약물에 대하여 부작용이 발생할 가능성을 유전체 정보를 통해 파악할 수 있다. 한 개인이 자신의 유전체 정보를 가지고 있다면, CYP2C19(약물 대사와 관련된 유전자)의 변이로 인해 Plavix와 같은 혈압 조절제를 복용하면 심각한 부작용이 나타나는 등의 일의 발생을 막을 수 있다. 혹은 CYP1A2(카페인 분해 능력 관련 유전자)의 변이가 있다면, 이로 인해 카페인을 먹은 후 가슴 두근거림 등의 부작용이 발생할 수 있다는 것을 알 수 있다. 이처럼 개인의 유전체 정보를 이용하면 특정 약물에 대한 부작용을 사전에 방지할 수 있다는 이점이 있다.

# Sequencing(염기 서열 결정)

## chain termination 방법

Sanger sequencing은 1977년에 Frederick Sanger에 의해 개발된 방법으로, 기본적으로 polymerase chain reaction(PCR) 과정에서 di-deoxynucleotide triphosphates(ddNTP)에 의해 DNA strand가 합성되지 않는 chain termination의 원리와, 크기에 따라 이동거리가 다른 양상을 보이는 전기영동의 원리를 이용하여 염기서열을 확인하는 방법이다. PCR 과정에서 DNA가 증폭되는 과정에서 dNTP 대신 ddNTP가 결합된다면, 3번 탄소에 수산화기 대신 수소가 붙어 있는 ddNTP로 인해 더이상 새로운 인산기가 결합될 수 없다. 따라서 증폭이 중지된다. 그러한 이유에서 chain termination이라고 부른다. 현재는 대규모 자동 게놈 분석을 위해 NGS 방법이 많이 쓰이지만, Sanger가 고안한 chain termination 방식은 여전히 유효하다. 그 이유는 NGS에 비해 작은 규모의 프로젝트에는 적용할 수 있다는 점, NGS보다 정확성이 높다는 점 때문이다. 다만 NGS에 비해 시간이 오래 걸릴 수 있다는 점이 trade-off다.

### chain termination의 원리

Sanger sequencing을 이해하기 위해서는 우선적으로 PCR 과정에 대해 이해할 필요가 있다. DNA 증폭을 위해서는 DNA를 구성하는 기본 단위인 deoxynucleotide triphosphate(dNTP)가 필요하다. 즉, DNA의 복제는 주형(template) DNA 가닥(strand)에 상보적인 dNTP가 결합하면서 이루어진다. DNA 서열에 dNTP가 결합될 때, DNA 서열의 끝 부분인, 마지막 nucleotide의 3번 탄소의 수산화기(-OH)와 새로운 dNTP 5번 탄소의 인산기가 반응하여 결합하게 된다. 이 과정에서 dNTP 대신 ddNTP를 이용하면(즉, DNA 복제 과정에서 DNA 가닥이 합성될 때 dNTP 대신 ddNTP가 결합하게 하면) DNA 복제가 중단된다. 앞서 언급했듯이 ddNTP에는 수산화기 대신 수소가 붙어 있기 때문에 반응에 참여할 산소가 없기 때문이다. 즉, Sanger sequencing 원리의 핵심은 ddNTP에 있다. 하지만 늘 DNA의 말단이 ddNTP와만 결합하는 것도 아니고 dNTP와만 결합하는 것도 아니다. 즉, '우연(random)'한 방식으로 dNTP에 결합할 수도 ddNTP에 결합할 수도 있다. 따라서 다양한 길이의 DNA 조각이 발생하게 된다. 즉, DNA 사슬(chain)의 중단(termination)이 발생하는 원리를 이용한 염기서열 분석법이다.

![](../images/2020-11-09-ngs/Untitled3.png)

#### ddNTP
dNTP는 3'-OH가 연결되어 있어 새로운 뉴클레오타이드의 인산이 탈수축합 결합을 형성하여 DNA가 복제될 수 있는 구조를 갖고 있다. 그러나 ddNTP는 dNTP의 3번 탄소에 산소(oxygen)가 존재하지 않는다. 즉 수산화기(-OH) 대신 수소(-H)가 결합되어 있는 nucleotide 구조이다. 따라서 다른 뉴클레오타이드와 결합할 수 없는 구조를 갖고 있다. DNA 복제가 더이상 일어날 수 없는 것이다. 

![](../images/2020-11-09-ngs/Untitled4.png)

#### chain termination을 이용한 염기서열의 확보
PCR이 완료된 후에는 전기영동을 통해 증폭된 DNA가 크기 순으로 나뉜다. 증폭된 각 DNA의 서열 끝에는 가장 마지막으로 결합된 서열이 위치할텐데, 그것이 바로 ddNTP가 결합한 부위이다. ddNTP가 결합한 이유로 더이상 증폭되지 못하고 끝난 것이다. 한편, ddNTP는 처음부터 형광물질(또는 방사성)으로 표지된 것을 사용한다. 따라서 형광물질의 종류에 따라 염기서열을 읽을 수 있다. 이 방법을 쓰기 위해 필요한 재료는 다음과 같다.

- 서열을 알아내려는 template DNA 단일 가닥(ssDNA)
- DNA primer
- DNA polymerase
- 네 개의 dNTP(dATP, dTTP, dGTP, dCTP)
- 서로 다른 색깔로 형광 표시된 4 종류의, 소량의 ddNTP (디디옥시뉴클레오티드;핵산별로 ddATP (아데닌), ddGTP (구아닌), ddTTP (티민), ddCTP (사이토신))

![](../images/2020-11-09-ngs/Untitled5.png)

![](../images/2020-11-09-ngs/Untitled6.png)

PCR이 완료된 후에는, 염기서열을 크기에 따라 작은 순서대로 정렬할 수 있는 전기영동을 이용하여 증폭된 서열을 정렬한다. 크기순으로 정렬된 서열의 순서대로 형광 신호를 인식시키면, 각 서열 끝에 마지막으로 결합된 ddNTP가 가지고 있던 형광물질 종류에 따라 서열을 읽을 수 있게 된다. 생어의 염기서열 분석법 이전에는 DNA를 길이 별로 정렬하는 것을 ‘수동’으로 진행했다. 그렇지만 생어는 전기영동을 활용함으로써, 크기순으로 자동으로 배열되도록 했다.

# NGS

- DNA 염기서열을 자동으로 읽어주는 장비들은 지난 수십 년 동안 기본적으로 노벨상 수상자인 프레드릭 생거(Frederick Sanger)가 개발한 방식이었다. 그래서 생거 시퀀서(Sanger Sequencer)로 분류된다. 생거 시퀀싱이란 A,T,G,C 염기의 분자량(molecular weight)의 차이를 전기영동 내린 후 레이저 빛을 각 전기영동 밴드에 쬐면 형광물질에 따라 특이적인 파장의 빛을 발하게 되며, 이를 순서대로 읽어내 각 염기를 읽어내는 방식이다. 가장 고전적인 방식이고 지금도 시퀀싱의 확증검사(validation test)로 사용되는 방법이다. 이러한 방식으로 인간 게놈 프로젝트가 수행됐고, 비록 2년 단축되긴 했지만 인간의 전체 게놈을 해독하는 데는 무려 13년이나 걸렸다.

- Sanger가 고안한 chain termination 방법의 한계점은 인간게놈프로젝트를 통해 나타나게 되었다. Human Genome은 단순히 전기영동으로 분석하기에는 그 양이 엄청나기 때문이다. chain termination 방법은 약 1000bp를 한번에 분석할 수 있다. 그러나 인간의 Genome은 3Bbs(30억bs)다. 따라서 대용량의 데이터를 빠르게 분석할 수 있는 기법의 필요성이 대두되었다. 그 결과 NGS와 같은 차세대 염기서열 분석법이 나타나게 되었다. 

- 2007년에 미국의 생명공학 기업가인 조너스 로스버그는 베일러 의과대학의 리처드 깁스 교수와 함께 자신의 454 라이프사이언스 회사를 통해 제임스 왓슨의 전체 유전자를 분석했다. 이 때 걸린 시간은 불과 13주였으며 비용 또한 100만 달러로 현저히 줄었다. 454라는 기계는 전체 유전자를 200여 개의 크기로 구성된 염기 조각으로 나누고, 잘게 나눈 각각의 조각을 읽어낸 후 그를 한 줄로 이어 원래 30억 개 규모의 전체 유전코드로 재구성하는 방식이다. 이런 방식은 생거(Sanger) 방식과 달리 대량의 병렬 데이터 생산이 가능했는데 이러한 시퀀서를 차세대 시퀀서(Next Generation Sequencer, NGS)라 부르며 454 기계가 최초의 NGS가 됐다.

- NGS의 등장은 모든 생명과학 연구자들과 의학 분야 연구자들에게 갑자기 엑솜, 유전체에 이르기까지 방대한 DNA 데이터를 다루게 했으며, NGS를 기반으로 한 유전체 분석법도 매년 바뀌고 있을 정도로 그 발전 속도가 빠르다. 2000년 초반 30억 달러를 들여 13년 동안 분석했던 시퀀싱 기술이 2007년 100만 달러에 4개월, 2011년에는 3천 달러에 고작 48시간 밖에 걸리지 않는 상용화된 NGS가 등장했다. 

- NGS를 이용하면 100만에서 10억 염기까지 한번에 분석할 수 있다. 사실 차세대 염기서열 분석법이 정확도 측면에서는 Sanger의 chain termination 방법보다 뒤떨어진다. 그럼에도 불구하고 빠른 시간 내에 대량의 DNA 데이터를 읽을 수 있다는 점이 NGS를 사용하는 이유이다. 인간게놈프로젝트는 13년에 걸쳐 무려 30조원의 비용이 들었다. 비용과 시간적인 측면에서 많은 양의 DNA 분석에 Sanger의 방법을 사용하기엔 무리라는 결론이 나온다. 반면, NGS는 100만원대로 빠른 분석이 가능하다. 그만큼 NGS는 대량의 DNA분석에 반드시 필요한 도구다. 현재는 Illumina사의 플랫폼이 가장 dominant하다. NGS에서는 기종에 따라 약간의 차이는 있지만 기본적으로 DNA 서열을 증폭하고, 그 후 형광 표식 등을 카메라로 찍은 후, 이미지 처리를 통해 서열을 확보하는 과정을 거친다.

- 반도체 기술의 압도적인 성능향상을 상징하는 무어의 법칙(Moore’s Law)에 견주어 볼 때, NGS가 등장한 2008년경부터는 아예 무어의 법칙을 능가하는 속도로 시퀀싱 비용이 떨어지고 분석 시간이 단축되기 시작했다. 사실은 IT 분야를 훨씬 뛰어넘는 압도적인 기술 혁신이 이 유전체 분석 분야에서 이뤄지고 있던 셈이다. 아직도 비용이 저렴하다고는 할 수 없지만 이러한 NGS의 발전이 시퀀싱 비용과 난이도를 계속 낮춰준 것은 사실이며, 이제 점점 더 중요해 지는 것은 시퀀싱의 데이터 생산이 아니라, 생산된 데이터에서 생물학적인 의미를 찾아낼 수 있는 분석이라는 점이 자명해졌다. 따라서 앞으로는 더욱 강력하고 고급스러운 바이오인포매틱스(bioinformatics) 분석에 상대적으로 더 많은 비용과 시간을 투입해야만 하는 시기를 맞이하게 될 것으로 보인다.

## NGS의 원리

NGS에는 여러 기술들이 있고 이에 따라 플랫폼도 다양하다. 그럼에도 불구하고 여러 방법들에서 공통적으로 나타나는 특징이 있다. 첫째로는 DNA 증폭이 필요하다는 점이다. DNA를 조각 내고, 각 조각들을 증폭하여 양을 늘리는 것이다. 두번째는, 증폭된 DNA를 특정한 어떤 곳에 ‘부착’한다는 점이다. 이 ‘부착’을 위해서, 증폭된 각 조각들은 특별히 만들어진 부착 서열과 결합하게 된다. 이렇게 부착을 통한 분석 방법의 큰 장점은 **방향성을 조절할 수 있다**는 것이다. 그래서 3’에서 5’ 방향과 5’에서 3’ 방향을 조절하여 양 방향의 DNA를 모두 읽어 낼 수 있다. 

### NGS의 기술별 분류

![](../images/2020-11-09-ngs/Untitled19.png)

NGS는 기술에 따라 크게 3가지로 구분할 수 있는데, 염기 서열 결정 방법에 따라 first generation, second generation, third generation으로 나눌 수 있다. 기본적으로 모든 시퀀서는 Sanger 방식 기반이었다. 그러나 1세대인 Sanger 방식과 달리 대량의 병렬 데이터 생산이 가능한 시퀀서가 등장했고 이를 차세대 시퀀서 (Next Generation Sequencer, 2nd Generation Sequencer)라고 한다. first generation이 바로 sanger의 방식이다. second generation과 third generation은 NGS에 해당하는데, 이는 해독되어지는 뉴클레오티드의 단위인 read의 길이에 따라 갈라진다. Illumina, Thermo Fisher 등이 short read를 해독할 수 있는 기술을 갖고 있다. PacBio와 Oxford Nanopore는 long read를 해독할 수 있는 기술을 갖고 있다. short read를 해독할 수 있는 second generation 방식은 다시 SBL과 SBS 방식으로 나누어진다. SBS 방식은 다시 CRT 방식과 SNA 방식으로 나누어진다. 염기 서열을 결정하기에 앞서, 해독하고자 하는 template DNA에 대한  증폭 과정을 거치게 되는데, 이를 template cluster generation이라고 한다. template cluster generation은 구슬(bead)방식, 유리판(solid-state) 방식으로 다시 나누어진다. 

최근에는 NGS를 넘어선 3세대 염기서열 분석법 또한 개발되고 있다. 이 3세대 염기서열 분석법은 DNA 증폭 과정을 생략하고 분석할 수 있도록 하는 기술로, 점점 상용화 단계에 접어들고 있다. 

### Pyrosequencing(또는 bead-based 방식)

- Pyrosequencing은 DNA 단일가닥을 비드(bead)에 부착하고, DNA 중합효소를 통해 반대쪽 가닥을 합성하면서 염기서열을 읽어나가는 방식이다. Pyrosequencing은 Pyro라는 이름에서 알 수 있듯이, 중합 시 발생하는 피로인산을 통해 염기서열을 알아내는 방식이다. 

- 구슬같이 생긴 nano 단위의 adopter를 기름방울(emulsion)에 가둬놓은 후, 이 emulsion 내에서 PCR 반응을 수행하여 증폭한다. 염기서열 결정을 위해 bead에 붙어있는 adopter와 template의 상보적인 서열을 바탕으로 결합시킨 후 증폭시키면, 단일 가닥의 증폭된 DNA를 확보하게 된다. 해당 기술은 Roche, Thermo Fisher 등이 가지고 있다. 방식은 다음과 같다. 
    - 먼저, dATP를 넣는다. 
    - 이때 만약 중합 되어야 할 염기가 아데닌이라면, 피로인산이 나오면서 DNA가 신장될 것이다. 이때 발생한 피로인산은 일련의 화학반응을 거치게 되어 빛을 내고, 이 빛을 감지하여 염기서열을 읽을 수 있는 상태가 된다. 
    - 만약 이때 빛이 나오지 않는다면 아데닌이 아니라는 것이고, 반응물을 씻어내고 구아닌, 티민, 사이토신에 대해서 같은 일을 반복한다.

위와 같은 방식으로 빛이 언제 발생하는지 알면 어떤 염기인지 알 수 있게 된다. 이런 Pyrosequencing의 장점으로는 한번에 긴 서열을 읽을 수 있다는 점이다. 그러나 단점은 반응물에 들어가는 비용이 크다는 것, 두번째로는 같은 염기가 6개 이상 반복될 경우 오차가 크게 발생한다는 점이다. 즉, 동일한 염기가 반복되는 경우 오차가 커진다.

![](../images/2020-11-09-ngs/Untitled20.png)

### Sequencing By Synthesis(SBS)

DNA 중합반응을 이용한 염기서열 분석법이다. 이 분석법은 어떤 면에서 Sanger의 분석법과 상당히 유사하다. 이 분석법에서는 조금 변형된 뉴클레오타이드를 사용하는데, 이 뉴클레오타이드는 DNA 분자에 연결될 때 형광이 발생하며, Teminator cap이라는 부위가 있어서 DNA가 더 이상 연장되지 못하도록 막는다. 이것은 ddNTP와 유사한 특징을 가진다. 다만 차이점은 ddNTP는 절대 연장되지 못하지만, Terminator cap의 경우 효소를 이용해 잘라낼 수 있기 때문에, 원하는 순간에 다시 DNA 연장을 재개할 수 있다. 이 분석법은 이렇게 변형된 뉴클레오타이드의 특성을 활용하여 염기서열을 알아내는 것이다. 네 종류의 뉴클레오타이드(A, T, G, C)는 각각 다른 형광을 띄고, 그러므로 한 염기가 결합했을 때 발생하는 형광의 종류를 분석하면 어떤 염기가 그 위치에 들어갔는지 알 수 있게 된다. 그 뒤, Teminator cap을 잘라내고 다음 염기를 중합하게 되면, 형광이 또 발생하고 이를 통해 그 다음 염기의 종류를 알아내고, 이 과정을 계속 반복하여 염기서열을 쭉 읽어낼 수 있다. 이 방식의 장점은 Pyrosequencing의 한계를 해결할 수 있다는 점이다. 즉, 동일한 염기가 반복되는 경우 오차가 커지는 문제를 해결할 수 있다. 그러나 이 방식은 긴 염기서열에 대해 큰 오차를 낸다는 문제를 갖는다.

### Sequencing By Ligation(SBL)

앞선 두 과정에서는 뉴클레오타이드를 DNA 중합효소를 이용하여 단일 가닥에서 이중 가닥으로 합성하는 과정을 사용하는데, 이 방법의 경우, 인위적으로 설계된 작은 DNA 조각을 DNA ligase로 연결하는 방식이다. 이 방법의 경우 굉장히 짧은 서열 밖에 읽지 못한다는 단점이 있다. 

### 세 방법 중 선택의 기준?

각각의 장단점이 있는 세가지 방법은 상황과 목적에 따라 다르게 쓰인다. 셋 중에 가장 널리 쓰이는 것은 두번째 방법인 Sequencing by synthesis(SBL)이다. 하지만, 읽어야하는 샘플의 크기에 따라 사용하는 방법은 달라진다.

## NGS 플랫폼

NGS 장비는 현재 일루미나(Illumina), 써모피셔사이언티픽(Thermo Fisher Scientific), 퍼시픽바이오사이언스(Pacific Biosciences), 옥스포드나노포어(Oxford Nanopore Technologies) 등의 회사에서 출시된 제품들이 관련 시장을 형성하고 있다. PCR 증폭(amplication) 방식에 따라 로슈(Roche, 454)와 써모피셔(구 Life Technologies)의 기종들은 유탁액 증폭(emulsion PCR) 방식으로, 일루미나(Illumina)의 기종들은 브릿지 증폭(Bridge PCR) 방식을 사용한다. 퍼시픽바이오사이언스, 옥스포드나노포어 등의 3세대 장비들은 2세대 장비의 짧은 분석범위의 단점을 보완한 것이다. 2세대 장비에서 가장 dominant하게 time을 consuming하는 PCR 증폭 과정을 없앤 덕분이다. 3세대 장비들은 DNA 단일분자를 real-time으로 읽을 수 있다. 또한 매우 긴 DNA 가닥의 염기서열을 한번에 읽을 수 있다는 장점이 있다. 그러나 개별 염기서열 측면에서 보면 정확도가 높지 않은 단점이 있다.

![](../images/2020-11-09-ngs/Untitled9.png)

![](../images/2020-11-09-ngs/Untitled10.png)

여러 장비 회사 중 Illumina의 NovaSeq이 현재로서는 가장 최신의 기기이다. 이 기기는 60명의 Human Genome을 단 하루만에 해독한다. 같은 회사 장비 중 HiSeq X는 하루에 한 명의 게놈만 해독 가능하다는 점에서 60배나 높은 성능을 보인다.

### 로슈(Roche)

2007년 로슈의 454 장비는 인류 역사상 최초의 NGS 플랫폼으로 첫 번째 대상자는 DNA 구조를 발견해 노벨상을 받은 제임스 왓슨이었다. 이어 2009년에는 'GS Junior'라는 소형장비를 전세계 최초로 내놓았고, 이 장비를 10시간 가동해서 대략 35MB의 데이터를 생산할 수 있었다. 또 다른 장비인 'GS FLX'는 한 번에 읽을 수 있는 길이(read length)가 700bp로 주니어(Junior) 모델보다 조금 길어 23시간 동안 700MB를 생산해낸다.

두 경우 모두 한 번 가동(run)으로 생산되는 데이터 양이 다른 기종에 비해 상대적으로 적어서 주로 파이로시퀀싱(Pyrosequencing) 방식으로 사용된다. 드노보시퀀싱(De novo sequencing), 홀지놈리시퀀싱(Whole genome resequencing), 타겟 DNA 영역(Target DNA region), 메타지노믹스(Metagenomics), 그리고 RNA 분석 등 다양한 응용 분야에 폭넓게 활용할 수 있으며 데이터를 저장하는 형식은 SFF 형태이다.

#### 로슈(Roche)의 454 시리즈
454 시리즈에서는 하나의 bead(구슬)에 하나의 DNA 사슬을 고정시킨 후, emPCR(emulsion PCR)로 증폭시킨다. 증폭이 끝나면 각 bead에는 수백만 개로 복제된, 동일한 DNA 조각들로 덮인 상태가 된다. 각 bead에는 PicoTiterPlate라는 장치에서 하나의 구멍 (well)에 하나씩 들어가서 시퀀싱에 개시된다. PicoTiterPlate는 테스트 튜브로 사용되는 소형의 다중 웰(well)이 있는 평판이다. 분석 연구에서 표준 도구인 MicroTiterPlate의 축소판이라고 할 수 있다. DNA 시퀀싱에 주로 사용된다. 454의 시퀀싱 기법은 Pyrosequencing chemistry라고 하며 Luciferase가 사용되어 염기의 빛을 측정하여 서열을 알아낼 수 있다.

![](../images/2020-11-09-ngs/Untitled11.png)

#### 로슈(Roche)의 454 시리즈의 강점
454 시리즈의 GS FLX+ system을 기준으로 read length가 600bp 또는 1,000bp에 이른다. 하나의 read에서 긴 염기를 읽어낼 수 있다는 점은 방대한 fragment를 모아 순서를 맞추는 어셈블리 작업, reference 서열에 맞추는 mapping 작업의 난이도를 낮출 수 있다것을 의미한다.

#### 로슈(Roche)의 454 시리즈의 한계
454에서는 한 번의 run에서 생산되는 데이터 양이 다른 기종에 비해서 상대적으로 적다. 채 GB단위가 되지 않는다. 또한, 시장에서 가장 dominant한 Illumina사의 NGS 데이터 형식과는 다른 .SFF 파일이다. 

### 일루미나 (Illumina)의 GA, HiSeq, MiSeq 시리즈

1998년 설립된 Solexa는 SBS(Sequence by Synthesis)라는 새로운 기술을 시퀀싱에 적용하였다. 2007년 일루미나는 Solexa를 합병한 후에 SBS기반의 다양한 NGS 장비를 발전시켜 왔다. 2012년 현재의 대표적인 기종인 HiSeq 2000의 경우, 2 * 100bp 리드(read)에서 11일까지의 런 타임(run time)으로 최대 540Gb에서 600Gb까지 데이터 생산이 가능하다. 최신 장비인 Hiseq 4000 경우는 이보다 훨씬 빠른 3.5일만에 30억 쌍 홀게놈 시퀀싱으로 12명을, 2일 만에 180개의 엑솜 또는 100개의 메신저 RNA 시퀀싱 데이터를 한 번에 생산할 수 있다. 2017년 초반에 출신된 차세대 장비인 노바식(NovaSeq)은 단 2일 만에 60명의 홀게놈 시퀀싱을 처리할 수 있는 성능(최대 6TB 데이터)을 낸다. 일루미나의 NGS 시리즈는 전반적으로 run 가격이 저렴하면서도 많은 데이터를 생산한다. 따라서 대용량의 시퀀싱을 대량으로 처리하는 연구에서 많이 사용된다.

#### 일루미나 (Illumina)사의 기술의 원리

일루미나(Illumina)사의 기술은 각기 다른 형광물질이 부착된 염기(dATP, dCTP, dTTP, dGTP)와 함께 DNA 중합효소(polymerase)를 넣어 주면, 각 DNA의 염기서열에 상보적인 염기가 합성/복제되면서 각기 다른 형광을 방출하게 되는 원리를 이용한 것이다. 이것을 장비에 부착된 CCD 카메라를 통해 사진으로 찍고, 장비는 형광이 발생된 클러스터의 위치를 기억하고 있다가 각 클러스터의 형광 변화를 감지하여 클러스터의 염기서열을 기록한다. 출시된 장비로는 iSeq, MiniSeq, MiSeq, NextSeq, HiSeq, NovaSeq 등이 있으며 열거된 순서에 따라 출력(output)이 높아져 대량의 데이터를 생산한다.

![](../images/2020-11-09-ngs/Untitled12.png)

일루미나의 NGS 기법에서는 브릿지 증폭(Bridge PCR)이라는 방식으로 슬라이드 위에 DNA 단편을 고정시킨 후에 최대 1,000 분자까지 증폭시켜, 같은 서열의 DNA 단편 집단을 형성시키는 게 특징읻. 일루미나에서는 이 집단을 클러스터(cluster)라고 표현하며 이 때 염기서열의 양끝에 붙은 어댑터(adaptor)가 flow cell 표면에 부착해 이를 브릿지(Bridge) 방식이라 한다. 즉, bead가 아닌 plate 위에서 bridge PCR 과정을 통해 DNA를 증폭시킨 후, 단일 가닥의 template만 확보하여 염기서열을 알아내는 방식이며 이를 solid-state based 방식이라고도 한다. 이 클러스터를 주형으로 네 종류 의 형광 표식 염기를 사용한 일 염기 합성반응인 SBS(Sequencing by synthesis)를 수행한다.

![](../images/2020-11-09-ngs/Untitled13.png)

![](../images/2020-11-09-ngs/Untitled14.png)

일루미나 시리즈에서는 DNA 서열을 다른 기법들처럼 용액(또는 Emultion) 안에서 증폭시키는 것이 아니라 판 위에 고정시킨 후에 판 위에서 구부러지면서 증폭되어 서열집단을 형성하는 것이 특징이다. 그렇게 형성된 클러스터라는 집단은 집단 별로 시퀀싱이 이루어져 각 read의 염기서열 정보로 전환되고 분석과정으로 넘어간다. 일루미나의 데이터 형식인 FASTQ 파일은 현재 가장 많은 분석 소프트웨어에서 널리 사용되고 있다.

2011년에는 소형장비인 마이식(MiSeq)도 선보였는데, MiSeq의 특징은 기존 대용량 시퀀서인 하이식(HiSeq)의 기법(chemistry)을 그대로 유지하면서 소형화에 성공해 장비의 크기와 가격을 줄이면서도 작업은 더 빠르고 간편하게 만들었다는 점이다. MiSeq은 2 x 150bp에서 최대 24시간의 run time으로 3.7-4.6Gb의 데이터를 생산하며, 향후 2 x 250bp가 되면 35시간 이상의 run time으로 7Gb까지 가능하다. 2012년에 발표된 HiSeq2500은 HiSeq2000을 업그레이드하여 MiSeq의 신속함을 겸비함으로써 27시간에 120Gb까지 데이터를 생산하는 기능도 추가로 갖추게 된다. 일루미나 시리즈의 데이터 저장 형태는 FastQ다.

### 라이프 테크놀로지스 (Life Technologies (ABI))의 SOLiD, 5500 시리즈

원래는 최초의 시퀀서를 등장시킨 어플라이드 바이오시스템즈(Applied Biosystems / ABI)라는 이름으로 알려져 있지만, 인비트로젠(Invitrogen)과 합병되어 현재는 라이프 테크놀로지스가 공식 명칭이 되었다. 라이프 테크놀로지스의 SOLiD 시리즈에서는 emulsion PCR 과정 이후의 Ligation을 사용한 시퀀싱이 특징이다. 이 기법에서는 간격을 두면서 두 개씩 염기를 읽는데, primer reset을 통해 독립적으로 다섯 번을 반복하기 때문에, 최종적으로는 각 염기를 두 번씩 중복하여 읽어서 정확도를 높인다. SOLiD 시리즈에서는 csfasta 형식으로 데이터가 생성된다.

![](../images/2020-11-09-ngs/Untitled15.png)

하나의 reed 위에 Primer(n)에서 Primer(n-4)까지 독립적으로 다섯 번, 염기들을 읽어낸다. 이는 그림의 검은 점에 해당되는 위치에 나타나있다. 다섯 번의 독립적인 사이클이 모두 완료되면 Read Position 1번에서 29번까지는 최종적으로 모두 두 개의 검은 점이 위치하게 되는데, 이는 두 번씩 독립적으로 읽힌다는 것을 의미한다.

SOLiD 후속 기종인 5500 시리즈에서는 ECC(Exact Call Chemistry) 기법이 적용되어 여섯 번째 프라이머가 추가되었다. 즉, 5500 시리즈에서는 정확도를 더 높이기 위해 상당수의 Read Position과 겹쳐지는 Primer가 추가되어, 염기들이 여러 번 읽인 후에 다시 한 번 더 독립적으로 읽어들여, 내부 검증을 강화시킨 방식이다. 이에 따라 read 상에서 상당수의 염기들이 서로 다른 프라이머로 읽히는데, 세 번씩 독립적으로 읽힌다. 이로써 정확도가 SOLiD 시리즈의 99.94%에서 99.99%로 향상되었다. 5500의 다른 특징은 여섯 개의 lane으로 구성된 FlowChip에서 각 lane 별로 서로 다른 시퀀싱을 동시에 수행하는 Pay-Per-Lane sequencing 방식도 가능하는 점이다. 때문에 시간과 비용을 절약할 수 있다. 5500 시리즈에서는 기본적으로 35, 60, 75bp 등의 read length로 XSQ (eXtensible SeQuence)라는 바이너리 파일 형식의 데이터가 생성된다.

![](../images/2020-11-09-ngs/Untitled16.png)

### 제3세대 시퀀서, 차차세대 시퀀서 (3rd Generation Seguencer, Next NGS)

- 제3세대 시퀀서는 시퀀싱 전의 PCR 증폭 과정이 생략되고, DNA 단일분자를 그대로 시퀀싱 할 수 있다는 점이 차세대 시퀀서와 구분되는 점이다. Single Molecule Real Time(SMRT) 기법이라고도 한다. 

- 시장에는 Pacific BioSciences사의 PacBio RS 시스템이 나와 있다. 이 기술은 염기서열을 결정할 때 한 번에 15Kbp이상을 읽을 수 있다. 이 기술의 특징은 해독하려는 DNA의 양쪽에 고리 모양의 adopter 서열을 양쪽 말단에 부착하는 것인데, 이렇게 완성된 형태를 SMRTbell이라고 한다. 이 SMRTbell을 ZMW(Zero Mode Well)이라 불리는 바닥에 고정된 nano 단위의 홀에 흘려주면, 이 작은 구멍으로 DNA 사슬이 통과하면서, DNA polymerase와 상보적으로 결합하는 염기의 형광을 인식하여 시퀀싱을 수행한다. 즉 ZMW 내에 존재하는 polymerase와 SMRTbell이 만나게 되고, 형광이 달린 dNTP와의 합성을 시작하게 된다. 이 때 합성 시 발생되는 형광을 검출해내는 방식인 것이다. 이 기술은 시퀀싱이 하루 안에 완료되며 최대 90Mb의 데이터를 생산하는데, read length가 평균 2,500-3,000bp로 매우 길다는 것이 장점이다.

![](../images/2020-11-09-ngs/Untitled26.png)

### 반도체 시퀀서 (Semiconductor Sequencer)

써모피셔의 대표적인 장비인 솔리드(SOLiD) 시리즈에서는 유탁액 증폭(emulsion PCR) 과정 이후의 결찰(Ligation)을 사용한 시퀀싱이 특징이다. 이 기법에서는 간격을 두면 서 두 개씩 염기를 읽는데, 프라이머 리셋(primer reset)을 통해 독립적으로 다섯 번을 반복하기 때문에, 최종적으로는 각 염기를 두 번씩 중복해서 읽어서 정확도를 높인다. 2011년에는 라이프 테크놀로지스에 소속된 아이온 토렌트(Ion Torrent)에서 반도체 기술 기반의 소형 시퀀서인 PGM(Personal Genome Machine)을 등장시켜 소형 시퀀서를 유행시켰다. 2012년에는 10Gb와 100Gb의 데이터를 생산하는 새로운 반도체 칩에 대응하는 아이온 프로톤(Ion Proton)을 발표했다. 이 장비는 미세한 비드(bead)에 DNA 라이브러리를 붙여 촘촘하게 짜여진 미세한 구멍(well)에 하나씩 들어가도록 하고 각 구멍 밑에는 전류를 측정할 수 있는 반도체 칩 회로가 설계되어 있어 상보적인 염기가 결합하는 합성 과정에서 방출되는 수소이온(H+)으로 인하여 pH가 낮아지는 것을 반도체칩의 센서가 탐지하고 분석하는 방식이다. 현재 출시된 장비로는 Ion Personal Genome Machine(PGM) 및 Ion S5 System 두 가지 종류가 있다. Ion PGM에서는 평균 100bp와 200bp의 read length로 SFF, FASTQ 형식 데이터를 생산하며, PGM에 부속된 서버에서 reference 서열에 대한 mapping도 수행되기 때문에 그 결과 파일인 BAM 파일도 함께 제공한다. 개발자인 로스버그(Rothberg) 박사가 454를 개발했던 장본인이기 때문에 기법이 454와 유사한 부분이 있지만, 기본적으로는 지금까지의 어느 NGS 장비와도 구별되는 반도체 시퀀싱이기 때문에 다른 NGS 시퀀싱 결과에 대한 상호검증에 효과적이다. 

![](../images/2020-11-09-ngs/Untitled17.png)

emulsion PCR로 DNA를 증폭하는 것은 다른 NGS 기종과 유사하지만, 다른 기법처럼 빛을 발생시키는 인공염기를 사용해서 빛을 인식하는 시퀀싱 방식이 아닌, 반도체 칩의 작은 구멍(well)에서 각각 DNA 합성이 일어나면서 나타나는 pH 변화를 직접 반도체에서 신호로 잡아내어 수행한다는 점이 특징이다.

![](../images/2020-11-09-ngs/Untitled18.png)

다른 NGS 장비에서 필요한 카메라와 빛 인식, 이미지 처리 등이 모두 생략되었기 때문에 장비가 작고 저렴해졌으며, 시퀀싱과 분석도 3시간 정도에 완료된다. 따라서 신속한 소규모 시퀀싱이 가장 큰 장점이다. 반도체 칩에서 실제 시퀀싱을 하기 때문에 세 종류의 칩 중에서 하나를 선택하면 데이터 생산량과 비용을 고려한 시퀀싱이 가능하다(10Mb to > 1Gb). 

### 옥스포드 나노포어 (Oxford Nanopore)

영국 회사인 '옥스포드 나노'는  2012년 2월 17일 플로리다주 마르코 아일랜드에서 개최된 '유전체생물학 및 기술 발전에 관한 모임(Advances in Genome Biology and Technology meeting)'에서 자사가 개발한 그리드온 시스템(GridION system)으로 분석한 DNA 시퀀싱 데이터를 처음으로 공개했다. 그런데 놀랍게도 15분 만에 시퀀싱이 되는 혁신적인 기술을 선보여 전세계를 깜짝 놀라게 했다. 주 제품인 미니온(MinION)은 시퀀서 역할을 하는 기기로 스마트폰보다 작은 크기의 초소형 장치다. 여기 끝에 USB가 달려 시퀀싱 결과가 15분 만에 바로 컴퓨터 화면에 보이는 놀라운 기술이다. 최근에는 15분만에 시퀀싱을 한 후 아이폰 같은 휴대폰에 바로 연결해서 분석 결과를 볼 수 있는 스미지온(SmidgION)의 실물이 공개됐다. 이런 신속한 결과 분석은 현장에서 바로 진단해 진료에 활용하는 포인트오브케어(Point of Care) 시대를 가져오게 할 것이다. 즉, 미지의 바이러스를 진료실 현장에서 검체를 받아 바로 시퀀싱 분석해서 원인을 파악하고 수술 전 특이 체질을 알아내어 사망의 위험을 감소시킬 수 있다. 비즈니스 세계에선 보다 광범위한 응용이 가능한데, 매장에서 화장품을 고르기 전 나의 피부 특성을 즉석에서 체크하고, 식당에서 음식을 먹기 전 맞춤 유전체 검사를 하는 날도 머지 않아 다가올 수 있다.

![](../images/2020-11-09-ngs/Untitled25.jpeg)

#### 나노포어 시퀀싱
- 나노포어 라는 것은 직경 nano(10억분의 1) meter 수준의 극소의 채널(구멍)을 DNA 단일 가닥이 통과하면서 형성되는 전위차를 검출하는 방식이다. 네 종류의 염기마다 서로 다른 세기의 전류가 흐르기 때문에 가능한 방식이다. 막 단백질의 바깥쪽에서 이중나선의 DNA를 DNA helicase 효소를 이용하여 단일 가닥으로 풀어주는데, 이 단일 가닥 DNA가 일정한 속도로 채널로 이동한다. 채널 내에 약 4bp의 염기가 들어갈 수 있는데, 이때 측정한 전위차를 통해 염기서열을 알아내게 된다. 채널 안에는 약 4개의 염기가 들어가므로 실제로 측정한 전위 차는 한 염기에 해당하는 것이 아닌, 4개의 염기에 의한 것이라는 점을 염두해야 한다.

![](../images/2020-11-09-ngs/Untitled27.png)

- NGS에서 염기서열을 읽어내기 위해서는 보통 DNA를 증폭하고 빛을 방출하는 인공염기를 사용해 복제할 필요가 있는데 그러한 조작과 시약이 모두 생략되는 셈이다. 참고로, 위에서 설명한 것처럼 제3세대에서는 DNA 증폭이, 반도체 시퀀싱에서는 인공염기가 이미 각각 생략되어 있다. 나노포어 시퀀싱이란, DNA 한 가닥을 생물학적 세공(biological pore) 속으로 통과시키면서 전기전도성의 차이를 측정해 다양한 염기를 판별하는 기술을 말한다. 

![](../images/2020-11-09-ngs/Untitled24.jpeg)

## NGS의 문제점
- NGS의 에러는 높은 편이다.
- Sequencing이 이루어지지 않는 영역이 존재할 가능성이 있다.
- 짧은 서열 길이에 따른 Alignment 오류 가능성이 있다.

### 문제점에 따른 해결책
- 동일한 위치를 반복적으로 읽는다. (아래의 number X 표시는 number번을 반복해서 읽는다는 것을 의미한다.)
    - Home SNP = 5X 이상
    - Hetero SNP = 30X 이상
- 사람의 게놈 길이는 약 30억 bp이다. 따라서 한 사람의 게놈을 평균 30번 해독해야 한 사람을 완전히 해독했다고 할 수 있다. 즉 90Gbp를 해독해야 한다. 이를 '생산량(해독 coverage)'이라고 하며 일르 '30x'와 같이 표현한다. 

![](../images/2020-11-09-ngs/Untitled28.png)

반복해서 읽을 경우 에러율이 낮아지는 것을 확인할 수 있다.

## Variation
보통은 아래와 같이 염기서열의 변이(mutation)을 찾아내는 데에 쓰인다.

![](../images/2020-11-09-ngs/Untitled30.png)

![](../images/2020-11-09-ngs/Untitled31.png)

- Genome Variation
    - Nucleotide Variation과 Structural Variation 두 가지로 나눌 수 있다. 
    - 일반적인 mutation은 Nucleotide Variation에 해당하며, SNP, InDel 등이 이에 속한다. 
    - 이 때엔 Reference genome을 기준으로 single nucleotide의 변이 또는 Insertion, Deletion 등을 검출해내게 된다.
- Structural Variation
    - 특정 유전자 구간, 1Kbp 이상의 구간이 지속적으로 증폭되는 CNV에 의해 copy number에 gain 또는 loss가 발생하는 것 또는,
    - 크기가 큰 염기서열들에 대해 Inversion이나 Translocation이 발생하는 것에 해당한다.

![](../images/2020-11-09-ngs/Untitled32.png)

게놈 내에는 단백질을 암호화하고 있는 영역인 Exon이 있다. Exome Sequencing의 대상이 되는 서열은 양이 적다. 그래서 100번 정도 반복하여 해독한다(100x).

### Clinical applications - Liquid biopsy

![](../images/2020-11-09-ngs/Untitled_1.png)

- NGS를 이용한 임상 진단 가운데 하나는 Liquid biopsy(액체 생검 검사; fluid biopsy 또는 fluid phase biopsy)다. Liquid biopsy는 혈액(blood), 타액(saliva), 소변(urine) 샘플을 통해 종양에 대한 다양한 정보를 발견할 수 있게 해 주는 방법이다.
- 혈액 또는 타액에는 다음의 물질들이 있다.
    - cancer detection 이 가능한 exosomes, microRNA (miRNA), circulating tumor cells, cancer antigens
- 혈액에는 암에서 유발된 두 가지 물질이 있다:
    - cfDNA(cell-free circulating tumor DNA 또는 circulating tumor DNA 또는 ctDNA; 암 세포에서 유래된 물질)
        - 암세포는 사멸할 때 ctDNA를 혈액속으로 방출한다. ctDNA상에서 발견되는 암 돌연변이 유전자는 전통적인 검사 방법으로 시행했을 때 발견되는 것들과 정확히 일치한다. ctDNA에 대해 NGS 또는 digital-PCR 방식을 통해 purifying 및 analyzing을 할 수 있다.
        - 특히 cfDNA는 plasma(혈장 또는 혈청)에서 주로 발견된다.
    - CTCs(Circulating Tumor Cells; 혈관 내에서 순환하는 암세포)
- 암 세포 외에 심장병 진단을 위해 다음의 물질을 검출하기도 한다.
    - CECs(Circulating Endothelial Cells)
- EDGC사(이원다이애그노믹스(주))의 경우, Liquid biopsy를 통해 cfDNA, CTCs로부터 암에 대한 조기진단(스크리닝), 재발 가능성에 대한 모니터링 및 항암제 효과 등을 분석한다. 이를 통해 검출된 암세포 유래 cfDNA를 NGS를 통해 해독한 후 bioinformatics 분석을 통해 암을 유발하는 유전적 변이가 있는지, 사용하려는 항암제에 대해 효과를 볼 수 있는지, 혈액 속에서 암을 유발하거나 암세포가 있는지 등을 확인할 수 있게 된다.
- 단지 혈액만으로 이러한 진단이 가능하다는 점에서 tumor tissue에 대한 조직 검사(Tissue biopsy)를 수행하지 않는 비침습적(noninvasive)인 방식이라고 할 수 있다. 암 검진과 진단 시 Tissue biopsy로 인한 환자의 burden을 줄이면서 암의 상태를 확인할 수 있는 것이다. 이는 NGS 기술의 발전 덕분이라고 할 수 있다. 또한 침습적인 방식에 비해 몇 달 혹은 몇 년 간의 종양의 변화 등을 detection할 수 있다는 점에서 유리하다. 다만, 아직까지는 Tissue biopsy 방식이 임상 진단에서는 gold standard(중요/유의미)하다고 할 수 있다.

![](../images/2020-11-09-ngs/Untitled_2.png)

- 위 그래프는 bioanalyzer로부터 나온 결과다(참고로, bioanalyzer는 DNA의 size(bp), conc.(ng/ul), molarity(nmol/l)에 대한 정보도 함께 내어준다). x축은 DNA base pair를, y축은 흡광도(intensity)라고 이해하면 된다. 즉, 특정 bp에서 강한 흡광을 보이면 해당 bp에서의 농도가 높다고 이해하면 된다.
- 그래프 상에서 유의미하게 보아야 할 곳은 peak 부분이다. peak가 뜬 위치를 통해 cfDNA의 발생과 그 농도를 파악할 수 있다. 암세포는 사멸할 때 cfDNA를 다량 내놓기 때문에 166~200bp에서 높은 intensity를 보인다면 암세포 사멸로 발생한 cfDNA라고 판단할 수 있다. 정상인에게서도 소량의 cfDNA가 발생하는데 그 농도는 1~10ng/ul, 즉 0.001~0.0001ng/ul이다. 위 그림에서는 0.004ng/ul의 cfDNA가 발견되었으므로 암세포 사멸에 의한 cfDNA가 다량 발견되고 있다고 해석 가능하다.

#### Tissue biopsy vs. Liquid biopsy

![](../images/2020-11-09-ngs/Untitled_3.png)

두 방식의 차이점을 비교한 table이다. 아직까지는 Tissue biopsy 방식이 gold standard이며, 양성과 음성을 모두 진단해낼 수 있고, 모든 종양에 대해 진단이 가능하다는 이점이 있다. 반면 Liquid biopsy 방식은 비침습적이고 저렴하며, 연속적으로 샘플링을 할 수 있다는 이점이 있으나, 진단 가능한 암의 종류가 한정적이며 그 외의 암에 대해서는 보조적인 수단으로만 활용할 수 있다는 한계가 있다. 또한 '암이 확실하게 없다'는 등의 음성에 대한 진단 여부는 알 수 없다는 단점도 있다. 또한, 초기 단계의 암을 detection 하기에는 아직 한계를 보이고 있다. 그 이유는 초기 단계일 수록 cfDNA가 매우 드물게 혈액 내에 존재하기 때문인데, 이를 해결하기 위해 60,000X 정도의 반복적인 NGS 분석을 거쳐야 한다.

![](../images/2020-11-09-ngs/Untitled_4.png)

![](../images/2020-11-09-ngs/Untitled_5.png)

Liquid biopsy를 이용하여 진단 가능한 암의 종류는 Lung Cancer, Breast Cancer, Colorectal Cancer(직장암), Head and Neck Cancer, Prostate Cancer(전립선암), Gastrointestinal Cancer(위암), Leukaemia(백혈병) 등이다.

### T cell

![](../images/2020-11-09-ngs/Untitled_6.png)

- T cell은 항원 특이적인 적응 면역(후천 면역 또는 후천성 면역 특이 면역 2차 방어작용 획득 면역)을 주관하는 림프구의 하나이다. 가슴샘(thymus)에서 성숙되기 때문에 첫글자를 따서 T세포라는 이름이 붙었다. 전체 림프구 중 약 4분의 3이 T세포이다. T세포는 아직 항원을 만나지 못한 미접촉 T세포와, 항원을 만나 성숙한 효과 T세포(보조 T세포, 세포독성 T세포, 자연살상 T세포), 그리고 기억 T세포로 분류된다.
- T cell은 혈액을 계속 돌아다니는데, 그 목표는 특정 표적(세균, 종양 세포)을 찾는데 있다. 표적을 발견하면 경보를 울려 다른 면역세포를 불러들인다.

#### immune checkpoint blockade

![](../images/2020-11-09-ngs/Untitled_7.png)

그러나 이러한 T cell의 경보는 immune checkpoint 단백질에 의해 무력화될 수 있다. checkpoint 단백질은 정상 세포의 표면에 존재하는데, 면역반응을 약화시켜 자가면역반응을 억제하여 조직을 보호하는 역할을 한다. 

![](../images/2020-11-09-ngs/Untitled_8.png)

암세포는 면역공격을 회피하기 위하여 면역기능을 억제하거나 T 세포면역관용(immunetolerance) 또는 면역편집(immuno-editing) 등을 통하여 면역회피(immune escape)를 추구한다. 이러한 회피 전략의 하나로써 면역체크포인트 기능을 변화시켜 T 세포의 기능을 억제한다. 즉 면역체크포인트는 암세포 파괴를 방해하는 단백질로 암은 억제 면역체크포인트를 활성화시켜 T 세포의 공격을 회피한다. 구체적으로, 암세포들은 checkpoint 분자(PD-L1)를 발현하여 T cell의 PD-1분자와 결합하여 정상적인 면역활동을 방해(blockade)한다. 즉, 이 둘이 결합하지 못하게 해야 T cell이 암세포를 공격할 수 있다. 

![](../images/2020-11-09-ngs/Untitled_9.png)

- CTLA-4 (Cytotoxic T lymphocyte associated antigen-4) 억제제
    - 면역관문억제제에는 CTLA-4 단클론항체로 ipilimumab가 있고, PD-1 단클론항체로 nivolumab, pembrolizumab 등이 있으며, PD-L1 단클론항체로 atezolizumab, avelumab, durvalumab 등이 있다.
- PD-1(Programmed Death-1) 억제제
    - 암세포는 PD-L1이라는 면역회피물질을 갖고 있어 면역세포를 무력화시키고 그 틈을 이용해 증식한다. 즉 암세포의 PD-L1과 T 세포의 PD-1이 결합하면 T 세포가 기능을 상실하고 사멸하게 된다. 하지만 투여된 PD-1 항체가 PD-L1과 PD-1의 결합부위 사이에 미리 결합하면 면역회피신호가 차단된다. 면역회피 신호를 받지 않은 T 세포는 암세포를 사멸한다.
- PD-L1(Programmed cell death-ligand 1) 억제제
    - 이 약제는 atezolizumab, durvalumab 등으로 체내 면역세포와 암세포에 발현하는 단백질인 PD-1/PD-L1 경로를 표적으로 하여 이들의 상호관계를 차단하면서 면역세포들이 암세포를 공격할 수 있도록 도우는 역할을 한다.

![](../images/2020-11-09-ngs/Untitled_10.png)

종양 세포를 기준으로 면역 세포들이 얼마나 있는지에 따라 Inflamed, Immune Excluded, Immune Desert로 나눠진다. checkpoint의 발현량을 측정하여 항암제의 잘 드는 정도를 알 수 있다. 이러한 항암제의 처방에는 NGS가 쓰인다. 다양한 항암제 조합이 가능하기 때문이다. 

![](../images/2020-11-09-ngs/Untitled_11.png)

- 면역 항암제의 종류와 방법은 다양하다.
    - TAA(Tumor Associated Anti-gen) vaccine
        - 종양 유래 vaccine 직접 투여
    - neoantigen vaccine
        - 종양 유래의 신생 항원을 백신으로 쓰는 방법
    - bulk TIL ACT
        - 종양 세포 주위에 몰려있는 면역 세포들을 activation 시키는 방법
    - checkpoint blockade 방법
    - neoantigen-targeted ACT
        - neoantigen을 targetting하여 activation 시키는 방법
- 이 방법들은 환자의 종양이 많은지 적은지에 따라 적용할지 말지가 결정된다. 예를 들어, 돌연변이가 많은 종양인 경우에는 neoantigen vaccine, checkpoint blockade 방법, neoantigen-targeted ACT 중에서 선택적으로 사용 가능하다.

#### 면역항암 치료에 NGS가 필요한 이유

환자 종양의 뮤테이션 여부, 그 정도 등을 확인하기 위해선 WGS(Whole Genome Sequencing)을 통해 RNA sequencing, Genomic DNA의 mutation burden이 높은지 아닌지를 확인하기 위한 WES(Whole Exome Sequencing)이 필요하다. 특히, 암의 종마다 돌연변이 율이 높은 암과 낮은 암이 다르다. 즉 WES, WGS를 통해 암 세포가 가진 돌연변이를 파악한다. 더불어, RNA-seq을 통해 돌연변이를 가진 암세포에 대한 항원의 전사체를 알아낸다. 이를 통해 해당 전사체가 translation된 것을 예측할 수 있게 된다. neoantigen을 예측할 수 있다면, 면역반응이 활성화되는지 테스트하는 등 더 적합한 치료방법을 적용할 수 있다. 

![](../images/2020-11-09-ngs/Untitled_12.png)

돌연변이가 가장 많은 암은 Melanoma이다. 돌연변이 가능성이 낮은 왼쪽의 암들은 targeted therapy를, 돌연변이 가능성이 높은 오른쪽의 암들은 immuno therapy를 적용한다. 돌연변이 가능성이 높다는 것은 어디에서 돌연변이가 발생할지 모르기 때문에 genome에 대한 상세한 분석이 필요하고, 이는 해당 유전자가 복잡할 수록 더 어려워진다.

![](../images/2020-11-09-ngs/Untitled_13.png)

면역항암제를 예측하는 분석 도구들은 위와 같다. NGS를 통해 해독된 데이터를 분석할 수 있는 툴들이다. 해독된 데이터를 위 도구들로 가공하면, 면역항암제를 어떻게 처방할지를 결정할 수 있게 된다.

![](../images/2020-11-09-ngs/Untitled_14.png)

신생 항원을 예측하기 위한 분석 툴 파이프라인은 위와 같다. Exome sequencing을 통해 돌연변이를 검출한 후, RNA 전사체 분석을 통해 돌연변이를 가진 전사체가 나오는지 파악하여, 궁극적으로는 돌연변이에 의해 변형된 peptide들을 예측해낸다.

##### TMB(Tumor Mutation Burden) panel target gene list

![](../images/2020-11-09-ngs/Untitled_15.png)

위 유전자들은 모두 종양에 관련된 유전자들이다. 암 환자가 가진 암에 대하여 이들의 변이를 파악하기 위해 사용된다. 이를 사람이 일일이 손으로 하지는 않는다. 즉, NGS를 통해 targetting하여 해독하는 작업을 거치게 된다.

### Big data applications

NGS를 통해 나오는 데이터는 빅데이터다. 이들 중 가장 많으면서도 잘 정리된 데이터가 암 데이터이다. 대부분의 암 관련 데이터는 TCGA(The Cancer Genome Atals)에 있다. 

![](../images/2020-11-09-ngs/Untitled_16.png)

TCGA는 암 조직과 암 주변 정상 조직에 대한 정보를 모아 놓은 데이터베이스다. 총 2.5PB의 크기로,  10종의 rare한 암을 포함한 33종의 암에 대한 데이터를 담고 있으며, 11,000명을 대상으로 한 데이터베이스라고 할 수 있다. 이를 통해 아래와 같이 유의미한 관련 논문들이 다수 나왔다. 

![](../images/2020-11-09-ngs/Untitled_17.png)

위 논문은 TCGA 데이터베이스 프로젝트가 끝난 후 나온 논문이다. 같은 Breast Cancer라 하더라도 subtype으로 나뉘어져 C1부터 C6까지로 나눠짐을 확인하였는데, 이는 유전체 데이터를 종합적으로 분석할 수 있었기에 가능한 결과라고 할 수 있다.

![](../images/2020-11-09-ngs/Untitled_18.png)

각 type별로 생존률 또한 알 수 있다. C3의 생존률이 가장 높다. 

![](../images/2020-11-09-ngs/Untitled_19.png)

TCGA 외에 다양한 데이터베이스에 이와 같은 유전체 정보들이 데이터베이스화 되고 있다. GTEx의 경우, 암이 아닌 다른 원인에 의해 사망한 데이털르 모아 놓은 데이터베이스다. 53개의 조직에 대한 정보를 보유하고 있다.

## other applications

NGS를 통해 RNA-seq, small RNA, miRNA를 분석할 수도 있다. 

![](../images/2020-11-09-ngs/Untitled_20.png)

위 그림은 종양 환자로부터 발생한 돌연변이의 정도를 나타낸 그래프다. 특정 유전자에 얼마만큼의 돌연변이가 발생했는지 NGS 데이터를 통해 확인할 수 있다.

![](../images/2020-11-09-ngs/Untitled_21.png)

![](../images/2020-11-09-ngs/Untitled_22.png)

![](../images/2020-11-09-ngs/Untitled_23.png)

산모의 혈액 속에서 검출되는 태아의 cfDNA를 통해 기형아 여부 등을 파악하는 데도 활용된다. 약 0.3X의 NGS 시퀀싱만으로도 오류 없이 검출해낸다. 

![](../images/2020-11-09-ngs/Untitled_24.png)

비만 예측 및 식습관 개선 등에도 활용한다. 이를 마이크로바이옴 또는 metagenome sequencing이라고 한다.
